```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank_img = np.zeros(shape = (512, 512, 3),dtype = np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
#Rectangle

new = cv2.rectangle(blank_img, pt1 =(50, 50), pt2 = (200, 150), color = (0, 255, 0), thickness = 5)
```

```python
plt.imshow(blank_img)
```

```python
#Square
new = cv2.rectangle(blank_img, pt1 =(200, 200), pt2 = (300, 300), color = (255, 0, 0), thickness = 5)
```

```python
plt.imshow(blank_img)
```

```python
#Circle
new = cv2.circle(blank_img, center =(400, 400), radius = 50, color = (0, 0, 255), thickness = -1)
```

```python
plt.imshow(blank_img)
```

```python
#Line
new = cv2.line(blank_img, pt1 =(0, 0), pt2 = (500, 500), color = (102, 255, 255), thickness = 5)
```

```python
plt.imshow(blank_img)
```

```python
#Text
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text = 'Hello', org = (10,500),fontFace = font, fontScale = 4, color = (255,255,255),thickness = 3, lineType = cv2.LINE_AA)
```

```python
plt.imshow(blank_img)
```

```python
new_img = np.zeros(shape = (512,512,3), dtype = np.int32)
```

```python
plt.imshow(new_img)
```

```python
#Polygons
```

```python
vertices = np.array([[100, 300],[200, 200],[400, 300],[200, 400]],dtype = np.int32)
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1, 1, 2))
```

```python
pts.shape
```

```python
cv2.polylines(new_img, [pts],isClosed = True, color = (255, 255, 255), thickness = 5)
```

```python
plt.imshow(new_img)
```
